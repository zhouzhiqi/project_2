# 特征工程

## 空白值增补

```python
# 医学上各个测量值不应出现0
# 推测0值代表病人未检测该项目
# 空白值均以median填补

def replace_nan(data):
    for col in data:
        data.loc[:,col] = data.loc[:,col].replace(np.nan, data.loc[:,col].median())
    return data

train = replace_nan(train)
test = replace_nan(test)

train.describe()
```

## 创建新特征

1.Pregnancies: 妊娠次数

    既往多妊娠的妇女,尤其是曾经分娩过巨大胎儿的妇女
    0: 0
    1: 1-3
    2: 3-9
    3: 9--
    
2.Glucose: 在口服葡萄糖耐量试验中血浆葡萄糖浓度为2小时

    正常值: 空腹血糖: 60~110 mg/dl; 餐后2h血糖 <140 mg/dl
    糖尿病: 空腹血糖: >=126 mg/dl; 餐后2h血糖 >=200 mg/dl
    0: 0-90
    1: 90-140
    2: 140-200
    3: >200
    不可能出现0

3.BloodPressure: 舒张压（mm Hg）

    低压, (收缩压为高压)
    高大,肥胖者血压较高,  男女不同, 与妊娠者分开
    正常值: 60-89, 高血压:>=90, 低血压:<60
    0: <60
    1: 60-89
    2: >90
    不可能出现0

4.SkinThickness: 皮褶厚度（mm）

    可以推断全身脂肪含量
    正常: <40,(中国)
    男女不同

5.Insulin: 2小时血清胰岛素（mu U / ml）

    糖尿病是由于胰腺不能产生足够的胰岛素
    或者身体的细胞对所产生的胰岛素没有适当的反应
    空腹者一般应介于5pmol/L到20pmol/L,餐后1小时,为空腹5～6倍；餐后2小时为空腹4～5倍；餐后3小时为空腹2.5～3倍.
    餐后2小时胰岛素正常值是不高于11.1mmol/l 大概不高于77.3mu U / ml

6.BMI: 体重指数（以千克/身高（m）为单位）^ 2）

    正常: <24 kg/m2
    超重: 24-28 kg/m2
    肥胖: >28 kg/m2

7.DiabetesPedigreeFunction: 糖尿病谱系功能

    糖尿病家庭病史
    看不懂, 以0.5分类吧
    0: <0.5
    1: 0.5-1.0
    2: 1.0-1.5
    3: >1.5

8.Age: 年龄（岁）

    1: 20-30
    2: 30-40
    3: 40-60
    4: 60-

9.Outcome: 类变量（0或1）

    1型: 大多数是在儿童中, 20岁以下
    2型: 最常见的糖尿病类型, 大多在40岁以上, 生活方式因素和遗传因素, 肥胖,缺乏体力活动,饮食不当,应力,和城市化, 
    妊娠糖尿病: 发生在所有怀孕的约2-10％，并可能在分娩后改善或消失 
    其他特定类型: 常染色体显性遗传形式, 前驱糖尿病, 成人隐匿性自身免疫性糖尿病成人隐匿性自身免疫性糖尿病

缺少属性值：是的


```python
def new_feature(data):
    data_cat = pd.DataFrame()

    data_cat["Pregnancies_cat"] = data["Pregnancies"]
    data_cat["Pregnancies_cat"].loc[(data_cat["Pregnancies_cat"]==0)] = 0
    data_cat["Pregnancies_cat"].loc[((data_cat["Pregnancies_cat"]>=1)&(data_cat["Pregnancies_cat"]<4))] =1
    data_cat["Pregnancies_cat"].loc[(data_cat["Pregnancies_cat"]>=4)&(data_cat["Pregnancies_cat"]<9)] =2
    data_cat["Pregnancies_cat"].loc[(data_cat["Pregnancies_cat"]>=9)] = 3


    data_cat["Glucose_cat"] = data["Glucose"]
    data_cat["Glucose_cat"].loc[data_cat["Glucose_cat"]<90] = 0
    data_cat["Glucose_cat"].loc[(data_cat["Glucose_cat"]>=90)&(data_cat["Glucose_cat"]<140)] =1
    data_cat["Glucose_cat"].loc[(data_cat["Glucose_cat"]>=140)&(data_cat["Glucose_cat"]<200)] =2
    data_cat["Glucose_cat"].loc[(data_cat["Glucose_cat"]>=200)] =3


    data_cat["BloodPressure_cat"] = data["BloodPressure"]
    data_cat["BloodPressure_cat"].loc[data_cat["BloodPressure_cat"]<60] = 0
    data_cat["BloodPressure_cat"].loc[(data_cat["BloodPressure_cat"]>=60)&(data_cat["BloodPressure_cat"]<90)] =1
    data_cat["BloodPressure_cat"].loc[(data_cat["BloodPressure_cat"]>=90)] =2

    data_cat["SkinThickness_cat"] = data["SkinThickness"]
    data_cat["SkinThickness_cat"].loc[data_cat["SkinThickness_cat"]<40] = 0
    data_cat["SkinThickness_cat"].loc[(data_cat["SkinThickness_cat"]>=40)] =1

    data_cat["BMI_cat"] = data["BMI"]
    data_cat["BMI_cat"].loc[data_cat["BMI_cat"]<24] = 0
    data_cat["BMI_cat"].loc[(data_cat["BMI_cat"]>=24)&(data_cat["BMI_cat"]<28)] =1
    data_cat["BMI_cat"].loc[(data_cat["BMI_cat"]>=28)] =2

    data_cat["DiabetesPedigreeFunction_cat"] = data["DiabetesPedigreeFunction"]
    data_cat["DiabetesPedigreeFunction_cat"].loc[data_cat["DiabetesPedigreeFunction_cat"]<0.5] = 0
    data_cat["DiabetesPedigreeFunction_cat"].loc[(data_cat["DiabetesPedigreeFunction_cat"]>=0.5)&(data_cat["DiabetesPedigreeFunction_cat"]<1)] =1
    data_cat["DiabetesPedigreeFunction_cat"].loc[(data_cat["DiabetesPedigreeFunction_cat"]>=1)&(data_cat["DiabetesPedigreeFunction_cat"]<1.5)] =1
    data_cat["DiabetesPedigreeFunction_cat"].loc[(data_cat["DiabetesPedigreeFunction_cat"]>=1.5)] =2

    data_cat["Age_cat"] = data["Age"]
    data_cat["Age_cat"].loc[(data_cat["Age_cat"]>=20)&(data_cat["Age_cat"]<30)] =1
    data_cat["Age_cat"].loc[(data_cat["Age_cat"]>=30)&(data_cat["Age_cat"]<40)] =2
    data_cat["Age_cat"].loc[(data_cat["Age_cat"]>=40)&(data_cat["Age_cat"]<60)] =3
    data_cat["Age_cat"].loc[(data_cat["Age_cat"]>=60)] =4
    
    return data_cat

new_feature_train = new_feature(train)
new_feature_test = new_feature(test)
```

```python
from sklearn.preprocessing import OneHotEncoder

encode = OneHotEncoder()
encode.fit(new_feature_train)

new_feature_train = encode.transform(new_feature_train).toarray()
new_feature_test = encode.transform(new_feature_test).toarray()
```

```python
def new(data):
    # 加强易导致肥胖的因素
    data["Insulin*Glucose"] = data["Insulin"] * data["Glucose"]
    data["SkinThickness*BMI"] = data["SkinThickness"] * data["BMI"]
    # 加强妊娠型糖尿病的因素
    data["Pregnancies**3"] = data["Pregnancies"] **3
    return data

train = new(train)
test = new(test)
test.head()
```
