## longitude&latitude

```python
#  避免出现极端情况, 取train['longitude']的 分为数在(1,99)
rang = (train['longitude']>np.percentile(train['longitude'],1))&(train['longitude']<np.percentile(train['longitude'],99))
sns.distplot(train.loc[rang,['longitude']], bins=50)
```

```python
#  避免出现极端情况, 取train['latitude']的 分位数在(1,99)
rang = (train['latitude']>np.percentile(train['latitude'],1)) & (train['latitude']<np.percentile(train['latitude'],99))
sns.distplot(train.loc[rang,['latitude']], bins=50)
```
